```python
#matplotlib
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

#numpy
import numpy as np
```

```python
#data acquisition
x_label, progressive_sum, progressive_error = np.loadtxt('../Data/02.1_integral_uniform_sampling.dat', usecols = (0,1,2), unpack = 'true')
x_label_samp, progressive_sum_samp, progressive_error_samp = np.loadtxt('../Data/02.1_integral_importance_sampling.dat', usecols = (0,1,2), unpack = 'true')

#style
plt.figure(figsize=(8,5))
plt.xlabel('N of throws')
plt.ylabel('$<r>-1$')
plt.grid(True)
plt.title('')

#plot
plt.errorbar(x_label,progressive_sum - 1.0, yerr = progressive_error, label = "Blocking average uniform sampling")
plt.errorbar(x_label_samp, progressive_sum_samp - 1.0, yerr = progressive_error_samp, label = "Blocking average importance sampling")
plt.axhline(0, ls = '-',c = 'red', label = "Exact value")

#legend
plt.legend()

plt.show()
```
