```python
#matplotlib
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

#numpy
import numpy as np

#center graphs
from IPython.core.display import HTML
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")
```

```python
#data acquisition
x_label, progressive_sum, progressive_error = np.loadtxt('../Data/03.1a_directly_call_option.dat',
                                                          unpack = True, usecols = (0, 1, 2))

#style
fig, axs = plt.subplots(1, 2, gridspec_kw={'width_ratios': [2, 1.5]}, figsize = (15, 5))

#plot 1
axs[0].errorbar(x_label,progressive_sum, yerr = progressive_error, label = "Blocking average")
axs[0].axhline(14.975790778311286, ls = '-', c = 'red', label = "Exact value")
#style plot 1
axs[0].set_xlabel('N of throws', fontsize = 15)
axs[0].set_ylabel('$<r>$', fontsize = 15)
axs[0].grid(True)
axs[0].set_title('Average value of r - blocking average', fontsize = 15)
axs[0].legend()

#plot 2
axs[1].plot(x_label, progressive_error, label = "Progressive uncertainty")
#style plot 2
axs[1].set_ylabel('$\sigma^2$', fontsize = 15)
axs[1].set_xlabel('N of throws', fontsize = 15)
axs[1].set_title('Average value of r - Progressive uncertainty', fontsize = 15)
axs[1].grid(True)

plt.show()
```
