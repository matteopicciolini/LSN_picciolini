```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

N = 100
x = np.arange(N)+1      # uso l'ordinamento da 1 a N

plt.figure(figsize = (12,10))
plt.suptitle("Distanza dal centro - Ground State", fontsize = 25)

#---------------------------

plt.subplot(2,1,1)

pos, ave, err = np.loadtxt("../Data/05.1_Unif_0.87_0.87_0.87_Dist_Ground_State.dat", usecols=(0, 1, 2), delimiter=' ', unpack = 'true')
plt.plot(x, pos, label = "unif")
pos, ave, err = np.loadtxt("../Data/05.1_Gauss_0.87_0.87_0.87_Dist_Ground_State.dat", usecols = (0, 1, 2), delimiter=' ', unpack = 'true')
plt.plot(x, pos, label = "gauss")
plt.ylabel('$d/a_0$', fontsize = 15)
plt.xlabel('Block number', fontsize = 15)
plt.grid(True)
plt.axhline(y=3/2, color = 'r', linestyle = '-')
plt.title('Media di blocco (= distanza istantanea)',  fontsize = 20)
plt.legend()

#---------------------------

plt.subplot(2,1,2)

pos, ave, err = np.loadtxt("../Data/05.1_Unif_0.87_0.87_0.87_Dist_Ground_State.dat", usecols = (0, 1, 2), unpack='true')
plt.errorbar(x, ave, yerr = err, label = "unif")
print('Risultato finale Unif: ')
print(ave[-1])
print('Incertezza statistica : ')
print(err[-1])

pos, ave, err = np.loadtxt("../Data/05.1_Gauss_0.87_0.87_0.87_Dist_Ground_State.dat", usecols = (0, 1, 2), unpack='true')
plt.errorbar(x, ave, yerr = err, label = "gauss")
print('\nRisultato finale Gauss: ')
print(ave[-1])
print('Incertezza statistica : ')
print(err[-1])

plt.ylabel(r'$\langle r\rangle /a_0$', fontsize = 15)
plt.xlabel('Block number', fontsize = 15)
plt.grid(True)
plt.axhline(y=3/2, color = 'r', linestyle = '-')
plt.title('Media progressiva',  fontsize = 20)
plt.legend()

plt.subplots_adjust(hspace=0.5)

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X,Y,Z = np.loadtxt("../Data/05.1_Unif_0.87_0.87_0.87_Ground_State.dat", usecols=(0,1,2), delimiter=' ', unpack='true')

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)

plt.show()

#--------------------------------------

# cross section on (x,y)=(0,0) axis

fig = plt.figure()
plt.hist(Z,200)
plt.xlabel("z")
plt.ylabel("Occurrences")
plt.title("Profilo verticale di densità di probabilità")

plt.show()
```
